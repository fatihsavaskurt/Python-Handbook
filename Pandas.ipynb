##### merge
-- İki df'i ortak bir anahtar sütun üzerinden birleştirmeye yarıyor. SQL'deki join mantığı aslında. <p>
-- En temel join türü INNER'dır. Her iki tabloda eşleşen kayıtları getirir. <p>

```python
df_merge1 = pd.read_csv("7-merge_data1.csv")
df_merge2 = pd.read_csv("7-merge_data2.csv")
```

```python
df_merge1
```

```python
df_merge2
```

```python
df_merge1.merge(df_merge2, on= "Employee_ID", how="inner") #eşleşenler birleşir.
```

```python
df_merge1.merge(df_merge2, on = "Employee_ID", how = "outer") # eşleşen birleşir, eşleşmeyen nan gelir.
```

```python
df_merge1.merge(df_merge2, on = "Employee_ID", how= "left") #sol tabloyu alır, sağdakiyle eşleştirir (eşleşenleri)
```

```python
df_merge1.merge(df_merge2, on = "Employee_ID", how= "right") #sağ tabloyu alır, soldakiyle eşleştirir (eşleşenleri)
```

#### apply

--Apply mantık olarak, benim verdiğim fonksiyonu dataframe içindeki parçalara uygula anlamına gelir. <p>

--Pandas içerisindeki mevcut fonksiyonlar da kullanılabilirken, kendimizin tanımladığı fonksiyonlar da df içine uygulanabilir. <p>

-- Daha çok bir kural seti üzerindne yeni bir kolon oluşturmak için kullanılır.

```python
df = pd.read_csv("8-apply_function_data.csv")
```

```python
"""
Case1
-----
Dataframe, salarye göre yüksek, orta ve düşük maaş kategorisi ekle. Fonksiyon yaz,
daha sonrasında bu fonkisyonu apply ile dataframe ekle.
"""
```

```python
df
```

```python
def salary_category (salary):
    if salary < 50000:
        return "Low"
    elif 50000 <= salary < 80000:
        return "Medium"
    else:
        return "High"
```

```python
df["Salary_Category"] = df["Salary"].apply(salary_category)
```

```python
df
```

```python
"""
Case2
-----
Experience > 10 --> performance_score + 1

İki yolu var:
1) Experience kolonu ile bir fonksiyon yazıp 10'dan fazla ise 1 değilse 0 döndür.
Daha sonrasında bu oluşturduğumuz 1-0'lı kolonu performance scorea ile toplatıp yeni bir 
kolon oluştur.
2) fonksiyon içerisinde row yazarak direkt olarak tüm rowları olaya dahil edip koşul
kontrolü yapmak

"""
```

```python
def ps_buff(experience):
    if experience > 10:
        return 1
    else:
        return 0
```

```python
df["Adjusted"] = df["Experience"].apply(ps_buff)
```

```python
df
```

```python
df["New Score"] = df["Performance_Score"] + df["Adjusted"]
```

```python
df
```

```python
def adjusted_func(row):
    if row["Experience"] > 10:
        return row["Performance_Score"] + 1
    else:
        return row["Performance_Score"]
```

```python
df["Adjusted_V2"] = df.apply(adjusted_func, axis=1)
```

```python
df["Formated_Name"] = df["Name"].apply(lambda x : x.replace("_", " "))
```

```python
df
```

```python
""" 
Yorumlar:

--En yüksek yaş ortalaması İzmir'de olmasına rağmen, bu tip yorumlarda
ortalama kullanılmamalı. Çünkü outlier değerler (aykırı değerler) olabilir. 
Yani yaşı 90 olan 2 kişi, yaşı 20 olan 5 kişilik bir grubun ortalamasını yukarı çeker.

--Bu tip analizlerde en mantıklı tespit median ile yapılır. Median ile, verisetinin
tam ortasındaki değerler baz alınır. Böylelikle outlier değerlerden kaçınılır.

Mean > Median → sağa çarpık dağılım

Mean < Median → sola çarpık dağılım

"""
```

```python
df.groupby("city")["age"].agg(["mean", "median", "std"])
```

```python
"""
Case 2 — Şehir Bazlı Satış Analizi

Senaryo

Her şehir için şunları bul:

Toplam sipariş sayısı

Ortalama sipariş tutarı

Toplam kazanç (sum)

"""
```

```python
df_nihai = df.groupby("city")["amount"].agg(
    siparis_sayisi = "count", 
    ortalama_tutar = "mean", 
    toplam_kazanc = "sum")
```

```python
df_nihai["siparis_basi_kazanc"] = df_nihai["siparis_sayisi"] / df_nihai["toplam_kazanc"]
```
