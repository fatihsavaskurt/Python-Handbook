```python
""" 
CASE 6 — Medal Analizi

Madalya alan sporcuları filtrele (dal boş olmMeayanlar)

Şu soruları cevapla:

En çok altın madalya alan ülke hangisi?

Toplam madalya sayısında ilk 5 ülke
"""
```

```python
""" 
CASE 7 — Matplotlib (ilk görselleştirme)

Yaş dağılımını görselleştir

En çok spor yapılan 10 branşı grafikle göster
"""
```

```python
df["Age"].unique()
```

```python
plt.hist(df["Age"])
plt.title("Yaş Dağılımı")
plt.xlabel("Yaş")
plt.ylabel("Adet")
plt.show()
```

```python
sport_count = df.groupby("Sport")["ID"].size().nlargest(10)
```

```python
sport_count
```

```python
plt.bar(sport_count.index, sport_count.values)
plt.title("En çok yapılan ilk 10 spor")
plt.xlabel("Spor")
plt.ylabel("Adet")
plt.show()
```

```python
def medal_flag(medal):
    if pd.notna(medal):
        return 1
    else:
        return 0
```

```python
df["medal_flag"] = df["Medal"].apply(medal_flag)
```
