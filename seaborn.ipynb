```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("athlete_events.csv")
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x = "Height", y = "Weight", data = df)
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.title("Athelete Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x = "Height", y = "Weight", data = df)
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.title("Athelete Weight vs Height")
plt.show()
```

### Hue
--Bununla neyi belirtmek istiyorsak onu belirtebiliyoruz. <p> 
--Bu örnekle beraber cinsiyetin boy ve kiloya göre dağılımını görmüş oluyoruz. <p>
--Çok fazla kategori içeren değişkenler bu örnek için mantıklı olmayabilir.

```python
sns.set_style("whitegrid")
sns.scatterplot(x = "Height", y = "Weight", hue="Sex", data = df)
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.title("Athelete Weight vs Height")
plt.show()
```

```python
df["Medal"].unique()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x = "Height", y = "Weight", hue="Medal", data = df)
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.title("Athelete Weight vs Height")
plt.show()
```

#### Size
--Ek bir kategori daha atanabiliyor, böylelikle örneğin cinsiyet kategorisinde, medal'in kırılımlara göre nerede olduğu grafikte gösterilebilir.

```python
sns.set_style("whitegrid")
sns.scatterplot(x = "Height", y = "Weight", hue="Sex", style="Medal", data = df)
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.title("Athelete Weight vs Height")
plt.show()
```

#### Style
--Bununla da büyüklük küçüklük ayarı kolona göre yapılabiliyor.

```python
sns.set_style("whitegrid")
sns.scatterplot(x = "Height", y = "Weight", hue="Sex", style="Medal", size = "Age", data = df)
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.title("Athelete Weight vs Height")
plt.show()
```

#### Boxplot Exercise
--Median, quartile, min-max 1.5 * IQR(25,50,75), outlier. Tek bir grafik içerisinde bunlar gözükecek. <p>
--Medyan ortadaki kutudaki çizgi yani 30,32 arasında. <p>
--Kutu datanın daha fazla dağıldığı <p>
--Yatay çizgiler max min kısımı <p>
--Outlier değer de en yukarıdaki daire

```python
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])

plt.figure(figsize=(6,5))
sns.boxplot(y=data)
plt.title("boxplot")
plt.grid(True)
```
