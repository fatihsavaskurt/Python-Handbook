```python
### Kolonların Numeric Hale Getirilmesi
```

```python
#Size

df_test = df.copy()
```

```python
df_test["Size"].unique()
```

```python
df_test["Size"]= df_test["Size"].str.replace('M', '000000')
df_test["Size"]= df_test["Size"].str.replace('k', '')
```

```python
df_test["Size"] = df_test["Size"].replace("Varies with device", np.nan)
```

```python
df_test[df_test["Size"] == "1,000+"]
```

```python
df_test = df_test.drop(index = 10472)
```

```python
df_test["Size"] = df_test["Size"].astype(float)
```

```python
df_test["Size"] = df_test["Size"] * 1_000_000
```

```python
df_test["Size"].unique()
```

```python
df_test.info()
```

```python
df_test["Installs"].unique()
```

```python
df_test["Price"].unique()
```

```python
remove_char = ['$', '+', ',']
add_columns = ['Installs', 'Price']
```

```python
for item in remove_char:
    for col in add_columns:
        df_test[col] = df_test[col].str.replace(item, "")
```

```python
df_test["Installs"].unique()
```

```python
df_test["Price"].unique()
```

```python
df_test["Installs"] = df_test["Installs"].astype(int)
```

```python
df_test["Price"] = df_test["Price"].astype(float)
```

```python
df_test.info()
```

```python
df_test["Installs"] = df_test["Installs"]/1000000000
```

```python
df_test.head(5)
```

```python
### Kolonların Date Hale Getirilmesi
```

```python
df_test["Last Updated"] = pd.to_datetime(df_test["Last Updated"], format="%B %d, %Y")
```

```python
df_test["Month"] = df_test["Last Updated"].dt.month
df_test["Day"] = df_test["Last Updated"].dt.day
df_test["Year"] = df_test["Last Updated"].dt.year
```

```python
df_test.head(5)
```

```python
""" 
Numeric kolonları incele:

dağılım (histogram)
outlier var mı?
skew var mı?
"""
```

```python
numeric_features = [feature for feature in df_test.columns if df_test[feature].dtype != 'O']
categorical_features = [feature for feature in df_test.columns if df_test[feature].dtype == 'O']
```

```python
plt.figure(figsize=(15,10))

for i in range(0, len(numeric_features)):
    plt.subplot(5,3, i+1)
    sns.kdeplot(x=df_test[numeric_features[i]])
    plt.tight_layout()
```

```python
plt.figure(figsize=(22,4))

category = ["Type", "Content Rating"]

for i in range(0, len(category)):
    plt.subplot(1,2, i+1)
    sns.countplot(x=df_test[category[i]], fill=True)
```

```python
#top 10 categories by installment
```

```python
top_installs = df_test.groupby(["Category"])["Installs"].sum().sort_values(ascending=False).reset_index().head(5)
```

```python
plt.figure(figsize=(15,4))
sns.barplot(x=top_installs["Category"], y= top_installs["Installs"]/1000000)
```

```python
#top 10 app by installment
```

```python
top_installs_app = df_test.groupby(["App"])["Installs"].sum().sort_values(ascending=False).reset_index().head(5)
```

```python
plt.figure(figsize= (15,10))

sns.barplot(x = top_installs_app["App"], y = top_installs_app["Installs"])
```
