```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
"""
Python'ın veri görselleştirme kütüphanesidir.
"""
```

```python
age_list = [10,20,30,30,30,40,50,60,70,75]
weight_list = [20,60,80,85,90,85,40,50,104,132]
```

```python
plt.plot(age_list, weight_list, "y") # y --> yellow, g, b, r
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
np_age_list = np.array(age_list)
np_weight_list = np.array(weight_list)
```

```python
plt.plot(np_age_list, np_weight_list, "r")
plt.xlabel("age")
plt.ylabel("weight")
plt.title("age vs weight")
plt.show()
```

```python
numpy_arr1 = np.linspace(0,10,20)
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1 ** 3
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1, numpy_arr2)
plt.show()
```

```python
numpy_arr3 = numpy_arr1 / 3
```

```python
plt.plot(numpy_arr1, numpy_arr3, "r*-")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1, numpy_arr2,"r")
plt.subplot(1,2,2)
plt.plot(numpy_arr2, numpy_arr1, "g")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.7,0.5])
figure_axes.plot(numpy_arr1, numpy_arr2, "g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_xlabel("Y Axis")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_fig = plt.figure(dpi = 100)
new_axes = new_fig.add_axes([0.1,0.1,0.8,0.8])
new_axes.plot(numpy_arr1**2, numpy_arr2*2, label = "numpy_arr1**2")
new_axes.legend()
plt.show()
```
