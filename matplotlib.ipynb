```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
"""
Python'ın veri görselleştirme kütüphanesidir.
"""
```

```python
age_list = [10,20,30,30,30,40,50,60,70,75]
weight_list = [20,60,80,85,90,85,40,50,104,132]
```

```python
plt.plot(age_list, weight_list, "y") # y --> yellow, g, b, r
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
np_age_list = np.array(age_list)
np_weight_list = np.array(weight_list)
```

```python
plt.plot(np_age_list, np_weight_list, "r")
plt.xlabel("age")
plt.ylabel("weight")
plt.title("age vs weight")
plt.show()
```
