```python
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
 'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
 'hire_date': [2004, 2008, 2012, 2014]})
```

```python
df1
```

```python
df2
```

```python
df3 = pd.merge(df1, df2) ## 2 farklı dataframein birleştirilmesi
```

```python
df3
```

```python
df4 = pd.DataFrame({'group': ['Accounting', 'Engineering', 'HR'],
 'supervisor': ['Carly', 'Guido', 'Steve']})
```

```python
df4
```

```python
df3
```

```python
df3[df3["employee"] == 'Bob']["group"].replace("Accounting", "Analyst", inplace= True)
```

```python
df3.loc[
    (df3["employee"] == 'Bob') & (df3["group"] == 'Accounting'),
    "group"
] = 'Analyst'
```

```python
df3
```

```python
df5 = pd.merge(df3, df4) ## 2 farklı dataframein birleştirilmesi
```
